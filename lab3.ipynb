```python
def sum_python(n):
 s = 0
 for i in range(n):
  s += i
  return s
print(sum_python(1000000))
%timeit sum_python(1000000)
```

```python
%load_ext cython
```

```python
%%cython
def sum_cython(int n):
    cdef long s = 0
    cdef int i
    for i in range(n):
        s += i
    return s
```

```python
print(sum_cython(1000000))
%timeit sum_cython(1000000)
```
